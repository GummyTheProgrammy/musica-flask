Análise de Vendas Fictícias
Neste notebook, vamos gerar um conjunto de dados aleatórios para simular vendas de produtos e, em seguida, fazer uma breve análise e visualização.

```python
# Célula 1: Configuração e Geração de Dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de estilo para os gráficos
sns.set_style("whitegrid")

# Geração de 100 linhas de dados fictícios
dados = {
    'Produto': np.random.choice(['Smartphone', 'Notebook', 'Tablet', 'Smartwatch'], 100),
    'Regiao': np.random.choice(['Norte', 'Sul', 'Leste', 'Oeste'], 100),
    'Vendas': np.random.randint(50, 500, 100)
}

# Cria um DataFrame do pandas a partir dos dados gerados
df_vendas = pd.DataFrame(dados)

# Mostra as 5 primeiras linhas do DataFrame para ver como ficou
df_vendas.head()
```

Agora que temos o nosso conjunto de dados, podemos fazer uma análise exploratória.
Vamos calcular algumas estatísticas básicas e agrupar os dados para entender as vendas por produto e por região.

```python
# Célula 2: Análise de Dados
print("--- Estatísticas Descritivas das Vendas ---")
print(df_vendas['Vendas'].describe())
print("\n" + "="*40 + "\n")

# Agrupa os dados para somar as vendas por produto
vendas_por_produto = df_vendas.groupby('Produto')['Vendas'].sum().reset_index()
print("--- Vendas Totais por Produto ---")
print(vendas_por_produto)
print("\n" + "="*40 + "\n")

# Agrupa os dados para somar as vendas por região
vendas_por_regiao = df_vendas.groupby('Regiao')['Vendas'].sum().reset_index()
print("--- Vendas Totais por Região ---")
print(vendas_por_regiao)
```

Para visualizar os resultados, vamos criar um gráfico de barras que mostra as vendas por produto.
Aqui, usamos a biblioteca seaborn para criar um gráfico de barras de forma simples e intuitiva. O gráfico será exibido logo abaixo da célula de código.

```python
# Célula 3: Visualização dos Dados
plt.figure(figsize=(10, 6))
sns.barplot(x='Produto', y='Vendas', data=vendas_por_produto)
plt.title('Vendas Totais por Produto')
plt.xlabel('Produto')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45) # Gira os rótulos do eixo X para melhor visualização
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Define a quantidade de colunas e linhas
num_colunas = 5
num_linhas = 20

# Cria nomes genéricos para as colunas
colunas = [f'Coluna {i+1}' for i in range(num_colunas)]

# Gera dados sintéticos para cada coluna
# Usando np.random.randint para criar números inteiros aleatórios
dados = {col: np.random.randint(0, 101, num_linhas) for col in colunas}

# Cria o DataFrame do pandas
df_generico = pd.DataFrame(dados)

# Mostra as primeiras 5 linhas para verificar
df_generico.head()
```

```python
import pandas as pd

# Assume que o DataFrame 'df_generico' foi criado na Célula 1
# e está disponível neste ambiente do Jupyter Notebook.

# Define o nome do arquivo Excel que será criado
nome_do_arquivo = 'relatorio_personalizado.xlsx'

# Usa um 'ExcelWriter' com o engine 'xlsxwriter'
with pd.ExcelWriter(nome_do_arquivo, engine='xlsxwriter') as writer:
    # Acessa o objeto 'workbook' e 'worksheet' do xlsxwriter para formatação
    workbook = writer.book
    worksheet = workbook.add_worksheet('Dados')

    # 1. Cria os estilos de formatação
    header_format = workbook.add_format({
        'bold': True,
        'font_color': '#FFFFFF',
        'bg_color': '#000000'
    })
    data_format = workbook.add_format({
        'font_color': '#FFFFFF',
        'bg_color': '#202022'
    })

    # 2. Escreve o cabeçalho e os dados com os estilos definidos
    for col_num, value in enumerate(df_generico.columns.values):
        worksheet.write(0, col_num, value, header_format)
    
    for row_num, row_data in enumerate(df_generico.values):
        worksheet.write_row(row_num + 1, 0, row_data, data_format)

    # 3. Adiciona as melhorias para o visual da planilha
    # Esconde todas as linhas de grade da planilha
    worksheet.hide_gridlines(2)
    
    # Auto-ajusta a largura das colunas para o conteúdo
    worksheet.autofit()
    
    # Define o zoom da planilha para 125%
    worksheet.set_zoom(125)

print(f"O arquivo '{nome_do_arquivo}' foi gerado com sucesso, sem as linhas de grade!")
```
